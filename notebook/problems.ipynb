## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

num_experiments = 1000

dice1 = np.random.randint(1, 7, num_experiments)
dice2 = np.random.randint(1, 7, num_experiments)
total_score = dice1 + dice2

condition_total_greater_than_7 = total_score > 7
count_total_greater_than_7 = np.sum(condition_total_greater_than_7)
probability_total_greater_than_7 = count_total_greater_than_7 / num_experiments

condition_total_even = total_score % 2 == 0
count_total_even = np.sum(condition_total_even)
probability_total_even = count_total_even / num_experiments

condition_both_conditions_met = condition_total_greater_than_7 & condition_total_even
count_both_conditions_met = np.sum(condition_both_conditions_met)

print(f"The probability that the total score is greater than 7 is {probability_total_greater_than_7:.3f}")
print(f"The probability that the total score is even is {probability_total_even:.3f}")
print(f"The number of times both conditions were met is {count_both_conditions_met}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np

ball_box = {}

""" # Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box) """

ball_box = ["White"] * 10 + ["Red"] * 20 + ["Green"] * 30

num_experiments = 1000

count_3_white_2_red = 0
count_all_same_color = 0

for _ in range(num_experiments):
    drawn_balls = np.random.choice(ball_box, size=5, replace=True)
    
    white_count = np.sum(drawn_balls == "White")
    red_count = np.sum(drawn_balls == "Red")
    green_count = np.sum(drawn_balls == "Green")

    count_3_white_2_red += int(white_count == 3 and red_count == 2)
    count_all_same_color += int(white_count == 5 or red_count == 5 or green_count == 5)

print(f"The probability of drawing exactly 3 white and 2 red balls is {(count_3_white_2_red / num_experiments) * 100:.2f}%")
print(f"The probability of drawing all balls of the same color is {(count_all_same_color / num_experiments) * 100:.2f}%")
```
